# Exploring the Data in Chunks

### How many columns have a numeric type? How many have a string type?

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', chunksize=3000)
num = []
obj = []
for chunk in chunk_iter:
    numeric_cols = [col for col in chunk.columns if chunk[col].dtype != 'O']
    num.append(len(numeric_cols))
    object_cols = [col for col in chunk.columns if chunk[col].dtype == 'O']
    obj.append(len(object_cols))
print('Number of num columns :', num)
print('Number of obj columns :', obj)
```

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', chunksize=3000)
obj=[]
i=0
for chunk in chunk_iter:
    i+=1
    object_cols = [col for col in chunk.columns if chunk[col].dtype == 'O']
    if len(obj)>0:
        if obj != object_cols:
            print('-'*10, ' Chunk_{} '.format(i),'-'*10)
            print(object_cols)
    else:
        obj = object_cols
        print(object_cols)  
    
```

### How many unique values are there in each string column?

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', chunksize=3000)
uniques = dict()
for chunk in chunk_iter:
    string_df = chunk.select_dtypes(include=['object'])
    string_col = string_df.columns
    for col in string_col:
        val_counts = string_df[col].value_counts()
        if col in uniques:
            uniques[col].append(val_counts)
        else:
            uniques[col] = [val_counts]

            
uniques_combined = {}

for col in uniques:
    u_concat = pd.concat(uniques[col])
    u_group = u_concat.groupby(u_concat.index).sum()
    uniques_combined[col] = u_group
    print(col, u_group.shape[0])
```

### Which float columns have missing values?

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', chunksize=3000)

missing = []
for chunk in chunk_iter:
    floats = chunk.select_dtypes(include=['float'])
    missing.append(floats.apply(pd.isnull).sum())

combined_missing = pd.concat(missing)
combined_missing.groupby(combined_missing.index).sum().sort_values()
```

### Calculate the total memory usage across all of the chunks.

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', chunksize=3000)
memory = 0
for chunk in chunk_iter:
    memory += chunk.memory_usage(deep=True).sum()/(1024*1024)
print('total memory is {}'.format(memory))
```

```python
convert_col_dtypes = {
    "sub_grade": "category", "home_ownership": "category", 
    "verification_status": "category", "purpose": "category"
}
date_cols = ["issue_d", "earliest_cr_line", "last_pymnt_d", "last_credit_pull_d"]
```

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', 
                         chunksize=3000, 
                         dtype=convert_col_dtypes, 
                         parse_dates=date_cols)
for chunk in chunk_iter:
    term_cleaned = chunk['term'].str.lstrip(" ").str.rstrip(" months")
    chunk['term'] = pd.to_numeric(term_cleaned)
    revol_cleaned = chunk['revol_util'].str.rstrip("%")
    chunk['revol_util'] = pd.to_numeric(revol_cleaned)
    
chunk.dtypes
```

```python
chunk_iter = pd.read_csv('OneDrive\Documents\my_datasets\Loans_2007.csv', 
                         chunksize=3000, 
                         dtype=convert_col_dtypes, 
                         parse_dates=date_cols)
memory = 0
for chunk in chunk_iter:
    term_cleaned = chunk['term'].str.lstrip(" ").str.rstrip(" months")
    chunk['term'] = pd.to_numeric(term_cleaned)
    revol_cleaned = chunk['revol_util'].str.rstrip("%")
    chunk['revol_util'] = pd.to_numeric(revol_cleaned)
    memory += chunk.memory_usage(deep=True).sum()/(1024*1024)
print('total memory is {}'.format(memory))
```

It looks like we were able to realize some powerful memory savings by converting to the category type and converting string columns to numeric ones.
